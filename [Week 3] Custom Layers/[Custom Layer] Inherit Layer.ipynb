```python
import tensorflow as tf
import numpy as np
```

```python
class SimpleDense(tf.keras.layers.Layer):
  def __init__(self, units = 32):
    super(SimpleDense, self).__init__()
    self.units = units
  def build(self, input_shape):
    w_init = tf.random_normal_initializer()
    print('the input shape is: ' + str(input_shape))
    self.w = tf.Variable(name = 'kernel', initial_value=w_init(shape = (input_shape[-1], self.units), dtype = 'float32'), trainable = True)
    b_init = tf.zeros_initializer()
    self.b = tf.Variable(name = 'bias',initial_value = b_init(shape = (self.units, ), dtype = 'float32'), trainable = True)
    super().build(input_shape)
  def call(self, inputs):
    return tf.matmul(inputs, self.w) + self.b
```

```python
layer = SimpleDense(units = 1)

var = tf.ones((1,1))
val = layer(var)
layer.variables
```

```python
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
_input = Input((1, ))
_output = SimpleDense(units = 1)(_input)
# simpleModel.add(tf.keras.layers.Dense(1, activation = 'relu', input_shape = (1, )))
# simpleModel.add(SimpleDense(units = 1))
simpleModel = Model(inputs = _input, outputs = _output)
simpleModel.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
simpleHistory = simpleModel.fit(xs, ys, epochs = 500, verbose = False)
```
