```python
import tensorflow as tf
import tensorflow.keras as K
```

```python
(x_train, y_train), (x_test, y_test) = K.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
class customModel(K.Model):
  def __init__(self):
    super().__init__()
    self.flatten1 = K.layers.Flatten()
    self.inner1 = K.layers.Dense(64, activation = 'relu')
    self.inner2 = K.layers.Dense(32, activation = 'relu')
    self.main_output = K.layers.Dense(10, activation = 'softmax')
  def call(self, inputs):
    input = inputs
    flatten = self.flatten1(input)
    inner1 = self.inner1(flatten)
    inner2 = self.inner2(inner1)
    return self.main_output(inner2)
```

```python
model = customModel()
model.compile(optimizer = 'sgd', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 10)
```

```python
model.evaluate(x_test, y_test)
```
